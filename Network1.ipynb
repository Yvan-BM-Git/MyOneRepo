```python
#Primero intalar en la ventana de comandos: pip3 install networkx
import networkx as nx
```

```python
G = nx.DiGraph() 
```

```python
nodos_G = ['a','b','c','d','e','f','g','h']
print(nodos_G)

G.add_nodes_from(nodos_G)
```

```python
G.order()
```

```python
lados_G=[('a','e'),('b','a'),('b','f'),('c','b'),('c','g'),('d','c'),('d','h'),('e','c'),('f','d'),('f','e'),('g','f'),('h','g')]
G.add_edges_from(lados_G)
```

```python
G.size()
```

```python
nx.degree(G,'c')
```

```python
list(G.degree())
```

```python
nx.draw_circular(G,with_labels=True,node_color='cyan')
```
